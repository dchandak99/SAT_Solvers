```python
def clause2list(expr):
    if z3.is_const(expr):
        return [str(expr)]
    elif z3.is_or(expr):
        return [atom for disjunct in expr.children()
                     for atom in clause2list(disjunct)]
    else:
        assert False, ('not supported', expr)

x, y, z = z3.Bools('x y z')
print(clause2list(z3.Or(x, y, z)))
#['x', 'y', 'z']
```
